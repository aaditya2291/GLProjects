## Q19. Plot a bar graph of Purpose vs Miles(Distance). (3 points)
### Note: Use the original dataframe without dropping the 'NA' values.
### Hint:You have to plot total/sum miles per purpose

```python
# creating a DataFrame of Purpose and Miles_per_Purpose and storing it in a separate variable to get the desired output
purpose_mean_miles = pd.DataFrame(df_uber_drivers.groupby("PURPOSE*")["MILES*"].sum().round())
print(purpose_mean_miles)

purpose_mean_miles.plot(kind="bar", color='Purple') # plotting a bar chart
plot.title('Purpose vs Miles/Purpose') # plot title 
plot.xlabel('Purpose') # x-axis label 
plot.ylabel('Miles by purpose') # Frequency miles
plot.legend() # showing legend 
plot.show() # function to show the plot
```

## Q21. Plot number of trips vs Category of trips. (3 points)
### Note: Use the original dataframe without dropping the 'NA' values.
### Hint : You can make a countplot or barplot.

```python
Cat_trip=df_uber_drivers.groupby('CATEGORY*')[['MILES*']].count()
fig, ax=plot.subplots()
ax.bar(Cat_trip.index,Cat_trip['MILES*'],color='orange')
labels =ax.get_xticklabels()
plot.setp(labels,rotation=45, horizontalalignment='right')
plot.title('No. of trips by Category')
plot.xlabel('Category')
plot.ylabel('No. of trips by category')
plot.show()
```

## Q22. What is proportion of miles that are covered as Business trips and what is the proportion of miles that are covered as Personal trips? (3 points)

### Note:Use the original dataframe without dropping the 'NA' values. The proportion calculation  is with respect to the 'miles' variable.
### Hint: Proportion is the miles covered as business trips by total sum of miles and miles  covered as personal trips by total number of miles

```python
round(pd.crosstab(index=df_uber_drivers['CATEGORY*'], columns=df_uber_drivers['CATEGORY*'], 
                  values=df_uber_drivers['MILES*'], aggfunc=sum, normalize=True, margins=True, margins_name="Proportion %")*100,(1))
```
